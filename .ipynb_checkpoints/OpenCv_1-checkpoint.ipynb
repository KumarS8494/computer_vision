<p style='color:black;font-weight:600;font-size:15px'>Converting into grayscale image:</p>

```python
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY) # Convert the color image to grayscale image
```

```python
plt.imshow(gray)
```




<p style='color:black;font-weight:600;font-size:15px'>Blurring the image::</p>

```python
blur = cv.GaussianBlur(gray,(19,19),0)
```

```python
plt.imshow(blur)
```

### Drawing with OpenCV:

<p style='color:black;font-weight:600;font-size:15px'>Drawing a line:</p>

<p style='color:black;font-size:20;font-weight:600;'>1.The image where the line should be drawn.</p>
<p style='color:black;font-size:20;font-weight:600;'>2.The starting point with both the x and y coordinates.</p>
<p style='color:black;font-size:20;font-weight:600;'>3.The ending point with both the x and y coordinates.</p>
<p style='color:black;font-size:20;font-weight:600;'>4.In the next property, we are assigning the color to the line. Here, the format is BGR. By utilizing the method that I have used we can get a blue line.</p>
<p style='color:black;font-size:20;font-weight:600;'>5.The last slot of attribute defines the thickness of the line.</p>
          

```python
#Drawing a Line
import numpy as np
```

```python
img = np.zeros((512,512,3))
```

```python
img = cv.line(img,(0,0), (511,511),(255,0.5,0.5),15)
```

```python
plt.imshow(img)
```

<p style='color:black;font-weight:600;font-size:15px'>Drawing a rectangle:</p>


```python
img2 = np.zeros((512,512,3))
```

```python
img2 = cv.rectangle(img2, (120,120),(400,400),(0,0,255),20)
```

```python
plt.imshow(img2)
```

<p style='color:black;font-weight:600;font-size:15px'>Drawing a circle:</p>

```python
img3 = np.zeros((512,512 ,3))
```

```python
img3 = cv.circle(img3,(155,355),100,(0,255,255),10)
```

```python
plt.imshow(img3)
```

<p style='color:black;font-weight:600;font-size:15px'>Inserting the text:</p>

```python
img = np.zeros((512,512,3))
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
```

```python
cv.putText(img, 'Computer Vision', (125,250), 
            font, 1, (0, 0, 255), 3, cv.LINE_AA)
```

```python
# cv.imshow("TEXT", img) # Display the TEXT
# cv.waitKey(0)
```

```python
plt.imshow(img)
```

<p style='color:black;font-weight:600;font-size:15px'>Drawing Polygons:</p>

```python
imgage1 = np.zeros((512,512,3))
```

```python
pts = np.array([[25,75],[25,160],
                [110,200],[200,160],
                [200,70],[110,20]],
               np.int32)
```

```python
pts = pts.reshape((-1,1,2))
```

```python
cv.polylines(image1,[pts],True,(0,255,255))
```

```python
plt.imshow(image1)
```

### 3. Accessing the Webcam :

```python
cap = cv.VideoCapture(0)
```

```python
while True:
    ret, frame = cap.read()
    plt.imshow(frame)
    cv.imshow('sk_Picture', frame)
    
    
    key = cv.waitKey(1)
    
    if(key == ord('q')):
        break
```

```python
cv.imwrite('test.png', frame)
```

```python
gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

```python
cv.imwrite('editPics.png', gray)
```

```python
blur = cv.GaussianBlur(gray,(19,19),0)
```

```python
plt.imshow(blur)
```

```python
cv.imwrite('blurPics.png', blur)
```
