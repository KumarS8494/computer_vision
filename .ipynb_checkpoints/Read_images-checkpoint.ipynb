<p style='color:Red; font-size:25px;'> NumPy : </p>

<p style='color:black;font-size:16px;'> NumPy’s main object is the homogeneous multidimensional array. It is a table of elements (usually numbers), all of the same type, indexed by a tuple of non-negative integers. In NumPy dimensions are called axes.</p>

<p style='color:black;font-size:16px;'>NumPy’s array class is called ndarray. It is also known by the alias array. Note that numpy.array is not the same as the Standard Python Library class array.array, which only handles one-dimensional arrays and offers less functionality. The more important attributes of an ndarray object are:</p>

<p style='color:red;font-size:16px;'>1. ndarray.ndim</p>
<p style='color:red;font-size:16px;'>2. ndarray.shape</p>
<p style='color:red;font-size:16px;'>3. ndarray.size</p>
<p style='color:red;font-size:16px;'>4. ndarray.dtype</p>
<p style='color:red;font-size:16px;'>5. ndarray.itemsize</p>
<p style='color:red;font-size:16px;'>6. ndarray.data</p>

```python
import numpy as np
```
