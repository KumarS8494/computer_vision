```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('image/img1.png')
```

```python
cv.imshow('image', img)
```

```python
plt.imshow(img)
```

```python
img = cv.line(img, (0,0),(255,255),(87, 245, 66), 20)
```

```python
plt.imshow(img)
```

```python
# cv.imshow("Lineimg", img)
```

```python
img=cv.arrowedLine(img, (0,255),(255,255),(87, 245, 66), 20)
```

```python
plt.imshow(img)
```

```python
img = cv.rectangle(img, (344,122),(542,222),(0,255,0),-1)
```

```python
plt.imshow(img)
```

```python
img = cv.circle(img, (344,233), 44, (255,0,0),-1)
```

```python
plt.imshow(img)
```

```python
font = cv.FONT_HERSHEY_TRIPLEX
```

```python
img = cv.putText(img, 'OpenCv FPIL', (10,300),font, 3,(0,255,255), 10, cv.LINE_AA)
```

```python
plt.imshow(img)
```
