### Histogram

```python
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = plt.imread('img1.png')
```

```python
plt.imshow(img)
```

<p style='color:back;font-weight:600;font-size:14px'>Convert Grayscaling</p>
<b>Importance of grayscaling</b>
<p style='color:gray;font-weight:600;font-size:14px'><b>Dimension reduction:</b> For example, In RGB images there are three color channels and has three dimensions while grayscale images are single-dimensional</p>
<p style='color:gray;font-weight:600;font-size:14px'><b>Reduces model complexity:</b> Consider training neural article on RGB images of 10x10x3 pixel. The input layer will have 300 input nodes. On the other hand, the same neural network will need only 100 input nodes for grayscale images.</p>
<p style='color:gray;font-weight:600;font-size:14px'><b>For other algorithms to work:</b> Many algorithms are customized to work only on grayscale images e.g. Canny edge detection function pre-implemented in OpenCV library works on Grayscale images only.</p>

```python
gray_image = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_image)
```

```python
plt.hist(gray_image.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k') #calculating histogram
```

```python
img = cv.imread('img1.png',0)
```

```python
hist = cv.calcHist([img],[0],None,[256],[0,256])
```

<p style='color:geen;font-size:14px;font-weight:400'>hist is a 256x1 array, each value corresponds to number of pixels in that image with its corresponding pixel value.</p>

##### Histogram Calculation in Numpy
<p style='color:geen;font-size:14px;font-weight:400'>Numpy also provides you a function, np.histogram(). So instead of calcHist() function, you can try below line : </p>

```python
hist,bins = np.histogram(img.ravel(),256,[0,256])
```

<p style='color:gray;font-size:14px;font-weight:400'>hist is same as we calculated before. But bins will have 257 elements, because Numpy calculates bins as 0-0.99, 1-1.99, 2-2.99 etc. So final range would be 255-255.99. To represent that, they also add 256 at end of bins. But we don't need that 256. Upto 255 is sufficient.</p>
<b>Note</b>
<p style='color:black;font-size:14px'>Numpy has another function, np.bincount() which is much faster than (around 10X) np.histogram(). So for one-dimensional histograms, you can better try that. Don't forget to set minlength = 256 in np.bincount. For example, hist = np.bincount(img.ravel(),minlength=256) OpenCV function is faster than (around 40X) than np.histogram(). So stick with OpenCV function.</p>
